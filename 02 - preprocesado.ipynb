```python
import pandas as pd
import numpy as np

z = pd.read_csv("train.csv")
print ("shape of loaded dataframe", z.shape)
```

```python
z.head()
```

```python
z.info()
```

```python
z.describe()
```

```python
# Verificar cuántos valores nulos hay en cada columna
z.isnull().sum()
```

```python
#Opereaciones de Limpieza

# Eliminar filas con datos faltantes
z.dropna(inplace=True)

print(z.isnull().sum())
```

```python
# Seleccionar columnas categóricas
categorical_cols = z.select_dtypes(include=['object']).columns

# Aplicar one-hot encoding a las columnas categóricas
z = pd.get_dummies(z, columns=categorical_cols, drop_first=True)

z.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
# Aplicar normalización a todas las columnas numéricas
z[z.select_dtypes(include=['float64', 'int64']).columns] = scaler.fit_transform(z.select_dtypes(include=['float64', 'int64']))
```

```python
# Guardar el DataFrame preprocesado
z.to_csv('train_preprocessed.csv', index=False)
```
